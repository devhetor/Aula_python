```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_olympics = pd.read_csv('athlete_events.csv', index_col=0)
```

```python
filtered_athlete = df_olympics[~df_olympics['Team'].str.contains('Brazil')]
brazil_team = filtered_athlete['Season'].value_counts().sort_index()
brazil_team.plot(kind='bar', color='green')
plt.title('Numero de times brasileiros por estação')
plt.xlabel('Estação')
plt.ylabel('Numero de times')
plt.show()
```

```python
team_count_by_sex = df_olympics['Team'].groupby(df_olympics['Sex']).count()
team_count_by_sex.plot(kind='bar', xlabel='Sexo', ylabel='Número de Times')
plt.show()
```

```python
team_count_by_sex = df_olympics['Team'].groupby(df_olympics['City']).count()
team_count_by_sex.plot(kind='bar', xlabel='Cidades', ylabel='Número de Times')
plt.show()
```

```python
df_olympics = df_olympics.sort_values(by=['Team', 'Age'])
first_30_teams = df_olympics['Team'].unique()[:30]
filtered_df = df_olympics[(df_olympics['Team'].isin(first_30_teams)) & (df_olympics['Age'] > 30)]
plt.figure(figsize=(10, 6))
plt.bar(filtered_df['Team'], filtered_df['Age'], color='green')
plt.xlabel('Times')
plt.ylabel('Idade')
plt.xticks(rotation=90)
plt.show()
```
